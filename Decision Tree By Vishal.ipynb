# Importing Data Set.

```python
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
"""
Importing data set 

"""

# Import
data = datasets.load_iris()
X = data.data
Y = data.target

# Verifying

print(data)
```

# classifying Decision Tree

```python
"""
Classifying Decision Tree

"""
classifying_tree = DecisionTreeClassifier(random_state=0)

# Text model of a tree
model = classifying_tree.fit(X,Y)
my_tree = tree.export_text(classifying_tree)
print(my_tree)
```

# Visualizing Decision Tree

```python

fig = plt.figure(figsize = (15,10))

vis_tree = tree.plot_tree(classifying_tree, feature_names = data.feature_names, class_names = data.target_names, filled = True)
```
